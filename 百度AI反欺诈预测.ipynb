```python
features = train_data.columns.tolist()
features.remove('label') # 去除label

# 特征的唯一值查看
for feature in features:
    print(feature, train_data[feature].nunique())
    
```

```python
train_data['fea_hash'].map(lambda x: len(str(x))).value_counts()
```

```python
remove_list = ['os','osv','lan', 'sid']
col = features
for c in remove_list:
    col.remove(c)
col    
```

```python
# 特征筛选
selected_feature = train_data[col]
# 构造一个新的特征 fea_hash_len
selected_feature['fea_hash_len'] = selected_feature['fea_hash'].map(lambda x: len(str(x)))
selected_feature['fea1_hash_len'] = selected_feature['fea1_hash'].map(lambda x: len(str(x)))
# Thinking 为什么要把很大的数字处理为0 异常值处理，长尾分布 
# 如果fea_hash很长的话，都归为0
selected_feature['fea_hash'] = selected_feature['fea_hash'].map(lambda x: 0 if len(str(x))> 16 else int(x))
selected_feature['fea1_hash'] = selected_feature['fea1_hash'].map(lambda x: 0 if len(str(x))> 16 else int(x))

selected_feature
```

```python
# test 特征筛选
test_selected_feature = test_1_data[col]
# 构造一个新的特征 fea_hash_len
test_selected_feature['fea_hash_len'] = test_selected_feature['fea_hash'].map(lambda x: len(str(x)))
test_selected_feature['fea1_hash_len'] = test_selected_feature['fea1_hash'].map(lambda x: len(str(x)))
# Thinking 为什么要把很大的数字处理为0 异常值处理，长尾分布 
# 如果fea_hash很长的话，都归为0
test_selected_feature['fea_hash'] = test_selected_feature['fea_hash'].map(lambda x: 0 if len(str(x))> 16 else int(x))
test_selected_feature['fea1_hash'] = selected_feature['fea1_hash'].map(lambda x: 0 if len(str(x))> 16 else int(x))

test_selected_feature.info()
```

```python
# 使用lgb 训练模型
import lightgbm as lgb

model = lgb.LGBMClassifier()
# 训练模型
model.fit(selected_feature.drop(['timestamp','version'],axis=1), train_data['label'])

# 模型预测
result = model.predict(test_selected_feature.drop(['timestamp','version'],axis=1))
result
```

```python
# submit 文件
res = pd.DataFrame(test_1_data['sid'])
res['label'] = result
res.to_csv("./track_baseline.csv", index=False)
```
